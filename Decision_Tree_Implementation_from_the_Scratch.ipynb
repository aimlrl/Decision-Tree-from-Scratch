```python
class TreeNode:

  def __init__(self,example_indices,left_child,right_child,criterion,
               unique_classes,class_probabilities,criterion_value,
               purest_feature_idx,division_mean):

    self.row_example_indices = example_indices
    self.left_node = left_child
    self.right_node = right_child
    self.node_loss_func = criterion
    self.unique_categories = unique_classes
    self.class_posterior = class_probabilities
    self.loss_func_value = criterion_value
    self.important_feature = purest_feature_idx
    self.cutpoint_value = division_mean
```

```python
class DecisionTreeClassifierRegressor:


  def __init__(self,criterion="self entropy",splitter="best",max_depth=None,
               min_samples_split=2,min_samples_leaf=1,
               min_weight_fraction_leaf=0.0,max_features=None,random_state=None,
               max_leaf_nodes=None,min_impurity_decrease=0.0,class_weight=None,
               ccp_alpha=0.0):
    
    self.loss_func = criterion
    self.split_stopping = splitter
    self.tree_max_depth = max_depth
    self.min_examples_split = min_samples_split
    self.min_examples_leaf = min_samples_leaf
    self.min_example_fraction = min_weight_fraction_leaf
    self.tree_features = max_features
    self.seed = random_state
    self.num_leaf_nodes = max_leaf_nodes
    self.impurity_decrease_thresh = min_impurity_decrease
    self.class_percentage = class_weight
    self.pruning_param_value = ccp_alpha


  def fit(self,X,y,sample_weight=None,check_input=True):
    np.random.seed(seed=self.seed)
    X_copy = np.copy(X)
    total_features = X_copy.shape[1]
    if self.tree_features == None:
      X_copy = X_copy[:,0:total_features]
    elif (self.tree_features == "sqrt") or (self.tree_features == "auto"):
      X_copy = X_copy[:,np.random.choice(total_features,size=total_features,replace=False)]
      X_copy = X_copy[:,0:int(np.sqrt(total_features))]
    elif self.tree_features == "log2":
      X_copy = X_copy[:,np.random.choice(total_features,size=total_features,replace=False)]
      X_copy = X_copy[:,0:int(np.log2(total_features))]
    y_copy = np.copy(y)
    y_copy = y_copy.reshape(y.shape[0],1)
    data = np.concatenate((X_copy,y_copy),axis=1)
    data_unique_categories = np.unique(data[:,data.shape[1]-1])
    example_idx = np.arange(0,data.shape[0])
    example_idx = example_idx.reshape(example_idx.shape[0],1)
    data = np.concatenate((data,example_idx),axis=1)
    example_idx = example_idx.reshape(example_idx.shape[0],)





    def loss_func_criterion_value(loss_func_criterion,divided_data):
      if loss_func_criterion == "self entropy":
        values, counts = np.unique(divided_data[:,divided_data.shape[1]-2],
                                 return_counts=True)
        class_proba_dict = dict(zip(values,(1/divided_data.shape[0])*counts))
        class_proba = np.array(list(class_proba_dict.values()))
        loss_func_value = -np.sum(class_proba*np.log(class_proba))
        return class_proba_dict, loss_func_value
      elif loss_func_criterion == "mse":
        return {}, np.mean((data[:,divided_data.shape[1]-2] - np.mean(data[:,divided_data.shape[1]-2]))**2)

    


    def determine_num_leaf_nodes(nodes_idx_list):
      nodes_indices_list = nodes_idx_list.copy()
      leaf_nodes_count = 0

      for i in range(len(nodes_indices_list)//2-1):
        if (nodes_indices_list[i] != None) and (nodes_indices_list[(2*i)+1] == None) and (nodes_indices_list[(2*i)+2] == None):
          leaf_nodes_count = leaf_nodes_count + 1

      return leaf_nodes_count




    def best_recursive_splitting(parent_node_indices,row_number,
                                 col_number,previous_left_node_loss_func_value,
                                 previous_right_node_loss_func_value,
                                 previous_loss_func_value,
                                 left_node_data_indices,
                                 right_node_data_indices,
                                 features_left_node_min_loss_func_value,
                                 features_right_node_min_loss_func_value,
                                 features_min_loss_func_value,
                                 features_cutpoint_values):
      
      if (col_number > (X_copy.shape[1]-1)) or (row_number > (X_copy.shape[0]-1)):
        return

      sorted_parent_node_data = data[X_copy[parent_node_indices,col_number].argsort()]
      #sorted_parent_node_data = sorted_parent_node_data.reshape(sorted_parent_node_data.shape[0],sorted_parent_node_data.shape[2])
      expected_left_node_data = sorted_parent_node_data[0:row_number,:]
      expected_right_node_data = sorted_parent_node_data[row_number:,:]
      if (expected_left_node_data.shape[0] > 0) and (expected_left_node_data.shape[0] <= parent_node_indices.shape[0]):
        left_node_class_proba_dict, left_node_loss_func_value = loss_func_criterion_value(self.loss_func,expected_left_node_data)
      else:
        left_node_class_proba_dict = {}
        left_node_loss_func_value = 0

      if (expected_right_node_data.shape[0] > 0) and (expected_right_node_data.shape[0] <= parent_node_indices.shape[0]):
        right_node_class_proba_dict, right_node_loss_func_value = loss_func_criterion_value(self.loss_func,expected_right_node_data)
      else:
        right_node_class_proba_dict = {}
        right_node_loss_func_value = 0

      if previous_loss_func_value > (left_node_loss_func_value + right_node_loss_func_value):
        previous_loss_func_value = (left_node_loss_func_value + right_node_loss_func_value)
        previous_left_node_loss_func_value = left_node_loss_func_value
        previous_right_node_loss_func_value = right_node_loss_func_value
        cutpoint_row_number = row_number
        features_min_loss_func_value[col_number] = previous_loss_func_value

        if cutpoint_row_number == 0:
          features_left_node_min_loss_func_value[col_number] = None
          left_node_data_indices[col_number] = None
          features_right_node_min_loss_func_value[col_number] = previous_right_node_loss_func_value
          right_node_data_indices[col_number] = sorted_parent_node_data[cutpoint_row_number:,data.shape[1]-1]
          features_cutpoint_values[col_number] = sorted_parent_node_data[cutpoint_row_number,col_number]

        elif cutpoint_row_number == (sorted_parent_node_data.shape[0]-1):
          features_left_node_min_loss_func_value[col_number] = previous_left_node_loss_func_value
          left_node_data_indices[col_number] = sorted_parent_node_data[0:cutpoint_row_number,data.shape[1]-1]
          features_right_node_min_loss_func_value[col_number] = None
          right_node_data_indices[col_number] = None
          features_cutpoint_values[col_number] = sorted_parent_node_data[cutpoint_row_number,col_number]

        else:
          features_left_node_min_loss_func_value[col_number] = previous_left_node_loss_func_value
          left_node_data_indices[col_number] = sorted_parent_node_data[0:cutpoint_row_number,data.shape[1]-1]
          features_right_node_min_loss_func_value[col_number] = previous_right_node_loss_func_value
          right_node_data_indices[col_number] = sorted_parent_node_data[cutpoint_row_number:,data.shape[1]-1]
          features_cutpoint_values[col_number] = (sorted_parent_node_data[cutpoint_row_number-1,col_number]+\
                                           sorted_parent_node_data[cutpoint_row_number,col_number])/2


      best_recursive_splitting(parent_node_indices,row_number+1,
                                 col_number,previous_left_node_loss_func_value,
                                 previous_right_node_loss_func_value,
                                 previous_loss_func_value,
                                 left_node_data_indices,
                                 right_node_data_indices,
                                 features_left_node_min_loss_func_value,
                                 features_right_node_min_loss_func_value,
                                 features_min_loss_func_value,
                                 features_cutpoint_values)
      
      row_number = 0
      previous_loss_func_value = np.inf
      


      best_recursive_splitting(parent_node_indices,row_number,col_number+1,
                               previous_loss_func_value,
                               left_node_data_indices,
                               right_node_data_indices,
                               previous_left_node_loss_func_value,
                               previous_right_node_loss_func_value,
                               previous_loss_func_value,
                               left_node_data_indices,
                               right_node_data_indices,
                               features_left_node_min_loss_func_value,
                               features_right_node_min_loss_func_value,
                               features_min_loss_func_value,
                               features_cutpoint_values)
      
      return  features_min_loss_func_value,
      features_left_node_min_loss_func_value,
      features_right_node_min_loss_func_value,
      left_node_data_indices,
      right_node_data_indices,
      features_cutpoint_values


    num_nodes = 0
    node_indices = []
    if self.split_stopping == "best":
      class_proba_dict, loss_func_value = loss_func_criterion_value(
        self.loss_func,data)
      root_node = TreeNode(example_indices=example_idx,left_child=None,
                         right_child=None,criterion=self.loss_func,
                         unique_classes=np.unique(y),
                         class_probabilities=class_proba_dict,
                         criterion_value=loss_func_value,purest_feature_idx=None,
                         division_mean=None)
      num_nodes = num_nodes + 1
      node_number = 0
      node_indices.append(node_number)
      
      bfs_traversed_tree = [root_node]

      while len(bfs_traversed_tree) > 0:
        if (np.floor(np.log2(len(node_indices))) <= self.tree_max_depth) and\
         (determine_num_leaf_nodes(node_indices) < self.num_leaf_nodes):

          if (bfs_traversed_tree[0].row_example_indices.shape[0] >= self.min_examples_split):
          
            features_min_loss_func_value,features_left_node_min_loss_func_value,features_right_node_min_loss_func_value,left_node_data_indices,right_node_data_indices,features_cutpoint_values = best_recursive_splitting(bfs_traversed_tree[0].row_example_indices,
                                                              0,0,np.inf,np.inf,np.inf,
                                                              [None]*X_copy.shape[1],
                                                              [None]*X_copy.shape[1],
                                                              np.zeros((X_copy.shape[1],)),
                                                              np.zeros((X_copy.shape[1],)),
                                                              np.zeros((X_copy.shape[1],)),
                                                              np.zeros((X_copy.shape[1],)))
          
            feature_idx = np.argmin(features_min_loss_func_value)
            bfs_traversed_tree[0].important_feature = feature_idx
            bfs_traversed_tree[0].cutpoint_value = features_cutpoint_values[feature_idx]

            if left_node_data_indices[feature_idx].size >= self.min_examples_leaf and\
           (abs(features_min_loss_func_value[feature_idx]-bfs_traversed_tree[0].loss_func_value) > self.impurity_decrease_thresh):

              data_indices = left_node_data_indices[feature_idx]
              class_proba_dict, loss_func_value = loss_func_criterion_value(
                  self.loss_func,data[data_indices])
              left_child_node = TreeNode(example_indices=data_indices,
                                      left_child=None,right_child=None,
                                      criterion=self.loss_func,
                                      unique_classes=y[data_indices],
                                      class_probabilities=class_proba_dict,
                                      criterion_value=loss_func_value,
                                      purest_feature_idx=None,division_mean=None)
              bfs_traversed_tree.append(left_child_node)
              num_nodes = num_nodes + 1
              node_number = node_number + 1
              node_indices.append(node_number)
            else:
              left_child_node = None
              node_indices.append(None)


            if right_node_data_indices[feature_idx].size >= self.min_examples_leaf and\
           (abs(features_min_loss_func_value[feature_idx]-bfs_traversed_tree[0].loss_func_value) > self.impurity_decrease_thresh):
              data_indices = right_node_data_indices[feature_idx]
              class_proba_dict, loss_func_value = loss_func_criterion_value(
                  self.loss_func,data[data_indices])
              right_child_node = TreeNode(example_indices=data_indices,
                                      left_child=None,right_child=None,
                                      criterion=self.loss_func,
                                      unique_classes=y[data_indices],
                                      class_probabilities=class_proba_dict,
                                      criterion_value=loss_func_value,
                                      purest_feature_idx=None,division_mean=None)
              bfs_traversed_tree.append(right_child_node)
              num_nodes = num_nodes + 1
              node_number = node_number + 1
              node_indices.append(node_number)
            else:
              right_child_node = None
              node_indices.append(None)

          else:
            left_child_node = None
            right_child_node = None

        bfs_traversed_tree[0].left_node = left_child_node
        bfs_traversed_tree[0].right_node = right_child_node

        bfs_traversed_tree.pop(0)
```
